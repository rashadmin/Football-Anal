```python
def season_unique_club(filename,season,all_seasons=False):
    df = pd.read_csv(filename).drop('Unnamed: 0',axis=1)
    if all_seasons:
        unique_team = df['season'].home_team.unique()
        return unique_team
    unique_team = df[df['season']==season].home_team.unique()
    return unique_team
```
